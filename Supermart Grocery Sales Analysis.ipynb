# 導入模組、資料

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib_inline
from datetime import datetime
 
df = pd.read_csv("Supermart Grocery Sales - Retail Analytics Dataset.csv")
df.head(20)
```

# 修正Order Date日期格式、新增Year欄位

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Order Year"] = df["Order Date"].dt.year
day=pd.to_datetime('2018/12/31')
df["Days_Recency"]=(day - df["Order Date"])
df.head(20)
```

# 資料觀察、處理

```python
df.describe()
df.info()
df.Region.value_counts()  
df[df['Region']=='North']
df[df['Customer Name']=='Harish']
```

```python
df['Order Year'].value_counts()
```

# 根據Sales計算折扣金額及原價

```python
df['Original Price'] = round((df['Sales'])/(1-(df['Discount'])),2)
df['Discounted Amount'] = df['Original Price'] - df['Sales']
df.head(2)
```

# 哪種產品類型銷售額在每年都位居前三

```python
sales_c = pd.pivot_table(data=df, index='Category', columns = 'Order Year', values='Sales', aggfunc='sum').reset_index()
sales_c['2015_Rank'] = sales_c[2015].rank(method='max', ascending=False)
sales_c['2016_Rank'] = sales_c[2016].rank(method='max', ascending=False)
sales_c['2017_Rank'] = sales_c[2017].rank(method='max', ascending=False)
sales_c['2018_Rank'] = sales_c[2018].rank(method='max', ascending=False)
print(sales_c)

sales_c[(sales_c['2015_Rank'] <=3) & (sales_c['2016_Rank'] <=3) & (sales_c['2017_Rank'] <=3) & (sales_c['2018_Rank'] <=3)]
```

# 各地區每一訂單平均銷售及利潤

```python
sales_orders_region = pd.pivot_table(data=df, index='Region', values=['Order ID', 'Sales', 'Profit'], 
               aggfunc = {'Order ID':'count', 'Sales':'sum', 'Profit':'sum'}).reset_index().sort_values(by='Sales', ascending=False)
sales_orders_region['Sales Per Order'] = sales_orders_region['Sales']/sales_orders_region['Order ID']
sales_orders_region['Profit Per Order'] = sales_orders_region['Profit']/sales_orders_region['Order ID']
sales_orders_region[['Region', 'Sales Per Order', 'Profit Per Order']]
```

# 各類別產品每一訂單平均銷售及利潤

```python
sales_orders_region = pd.pivot_table(data=df, index='Category', values=['Order ID', 'Sales', 'Profit'], 
               aggfunc = {'Order ID':'count', 'Sales':'sum', 'Profit':'sum'}).reset_index().sort_values(by='Sales', ascending=False)
sales_orders_region['Sales Per Order'] = sales_orders_region['Sales']/sales_orders_region['Order ID']
sales_orders_region['Profit Per Order'] = sales_orders_region['Profit']/sales_orders_region['Order ID']
sales_orders_region[['Category', 'Sales Per Order', 'Profit Per Order']]
```

# 每一類別產品每一銷售所得利潤

```python
profit = pd.pivot_table(data=df, index='Category', values=['Sales', 'Profit'], aggfunc='sum').reset_index()
profit['Profit Margin'] = round((profit['Profit']/profit['Sales']),4)
profit.sort_values(by='Profit Margin', ascending=False, inplace = True)
profit
```

```python
profit = pd.pivot_table(data=df, index='Sub Category', values=['Sales', 'Profit'], aggfunc='sum').reset_index()
profit['Profit Margin'] = round((profit['Profit']/profit['Sales']),4)
profit.sort_values(by='Profit Margin', ascending=False, inplace = True)
profit
```

# 最有價值顧客

```python
customers = pd.pivot_table(data=df, index=['Customer Name'], columns='Order Year',
                           values=['Order ID','Sales',"Days_Recency"], aggfunc={'Order ID':'count','Sales':'mean',"Days_Recency":'min'})
customers['Customer Value_2015']=customers['Order ID'][2015]*customers['Sales'][2015]
customers['Customer Value_2016']=customers['Order ID'][2016]*customers['Sales'][2016]
customers['Customer Value_2017']=customers['Order ID'][2017]*customers['Sales'][2017]
customers['Customer Value_2018']=customers['Order ID'][2018]*customers['Sales'][2018]
customers['2015_Rank'] = customers['Customer Value_2015'].rank(method='max', ascending=False)
customers['2016_Rank'] = customers['Customer Value_2016'].rank(method='max', ascending=False)
customers['2017_Rank'] = customers['Customer Value_2017'].rank(method='max', ascending=False)
customers['2018_Rank'] = customers['Customer Value_2018'].rank(method='max', ascending=False)
customers['Days_Recency_2018_Rank']=customers['Days_Recency'][2018].rank()
#customer

Best_Value_customers = customers[(customers['2015_Rank']<30) & (customers['2016_Rank']<30) & (customers['2017_Rank']<30) & (customers['2018_Rank']<30)]
Best_Value_customers
#df['Customer Name'].unique()
customers
```
